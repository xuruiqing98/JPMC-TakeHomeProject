```python
import json
import matplotlib.pyplot as plt

# 读取你的 metrics 文件
with open("../results/model_metrics.json", "r", encoding="utf-8") as f:
    metrics = json.load(f)

# 找到最佳模型（你代码里已经选好了）
best_model = metrics["best_model"]

# 拿到 threshold 对应的结果
threshold_info = metrics["validation_results"][best_model]["threshold_sensitivity"]

# 转成 list
thresholds = [float(t) for t in threshold_info.keys()]
precisions = [threshold_info[str(t)]["precision"] for t in thresholds]
recalls = [threshold_info[str(t)]["recall"] for t in thresholds]

# 画图
plt.figure(figsize=(6, 5))
plt.plot(thresholds, precisions, marker="o", label="Precision")
plt.plot(thresholds, recalls, marker="o", label="Recall")

plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Precision-Recall Tradeoff Across Thresholds")
plt.legend()
plt.tight_layout()
plt.show()
```
